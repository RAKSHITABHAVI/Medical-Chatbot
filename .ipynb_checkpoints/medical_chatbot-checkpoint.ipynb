```python
import pandas as pd
from rapidfuzz import process

# Load CSV
faq_df = pd.read_csv("medical_faq.csv")

def ask_medical_bot(question):
    questions = faq_df['question'].tolist()
    match, score, _ = process.extractOne(question, questions)
    if score > 50:
        return faq_df[faq_df['question'] == match]['answer'].values[0]
    else:
        return "Sorry, I don't know the answer. Please consult a doctor."
```
